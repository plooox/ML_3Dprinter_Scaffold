```python
# Test parameter setting
arr = np.empty((0, 3), int)
for bed_degree in range(55, 86, 3):
    for print_degree in range(180, 241, 3):
        for print_speed in range(30, 151, 3):
            arr = np.append(arr, np.array(
                [[bed_degree, print_degree, print_speed]]), axis=0)

nine_data = pd.read_csv('Data_Summary.csv')

dataset = nine_data.copy()
dataset = dataset.astype({'fidelity': 'float32'})

X = pd.DataFrame(dataset, columns=[
                 'bed_degree', 'print_degree', 'print_speed'])
Y = pd.DataFrame(dataset, columns=['fidelity'])
stdd = pd.DataFrame(dataset, columns=['std'])
summ = pd.DataFrame(dataset, columns=['sum'])
normalizer = StandardScaler()

stdd = normalizer.fit_transform(stdd)
summ = normalizer.fit_transform(summ)

class_array = np.empty((0, 1), int)
```

```python
# 9 Divide
for i in range(len(stdd)):
    if(summ[i] >= 0.75):
        if(stdd[i] < 0.25):
            class_array = np.vstack((class_array, np.array([1])))
        elif(stdd[i] < 0.75 and stdd[i] >= 0.25):
            class_array = np.vstack((class_array, np.array([2])))
        else:
            class_array = np.vstack((class_array, np.array([4])))
    elif(summ[i] >= 0.25 and summ[i] < 0.75):
        if(stdd[i] < 0.25):
            class_array = np.vstack((class_array, np.array([3])))
        elif(stdd[i] < 0.75 and stdd[i] >= 0.25):
            class_array = np.vstack((class_array, np.array([5])))
        else:
            class_array = np.vstack((class_array, np.array([7])))
    else:
        if(stdd[i] < 0.25):
            class_array = np.vstack((class_array, np.array([6])))
        elif(stdd[i] < 0.75 and stdd[i] >= 0.25):
            class_array = np.vstack((class_array, np.array([8])))
        else:
            class_array = np.vstack((class_array, np.array([9])))
```
